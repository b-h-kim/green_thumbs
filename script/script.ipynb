```python
url = f'https://plants.ces.ncsu.edu/plants/lavandula-angustifolia/'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
def get_dl(soup):
    keys, values = [], []
    for dl in soup.find('div', attrs = 'bricks').findAll("dl"):
        for dt in dl.findAll("dt"):
            keys.append(dt.text.strip())
        for dd in dl.findAll("dd"):
            values.append(dd.text.strip())
    return dict(zip(keys, values))

dl_dict = get_dl(soup)
```
